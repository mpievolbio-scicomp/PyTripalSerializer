# Playground for dask distributed
https://distributed.dask.org/en/latest/quickstart.html

In terminal and tripser env activated, run `dask scheduler`

```python
from dask.distributed import Client
```

```python
client = Client('127.0.0.1:8786')
```

```python
client
```

```python
def square(x):
        return x ** 2
```

```python
A = client.map(square, range(10))
```

```python
client.gather(A)
```

```python
sum_of_squares = client.submit(sum, A)
```

```python
sum_of_squares.result()
```

```python
client.restart()
```

```python
squares = client.gather(A)
```
